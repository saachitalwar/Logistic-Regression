```python
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
```

```python
train_data = pd.read_csv("C:/Users/Saachi/Desktop/CN/clusterfuck/titanic_train.csv",delimiter=",")
train_data
```

```python
train_data.isnull().sum()
```

```python
train = train_data.copy()
train['Age'].fillna(train['Age'].median(skipna=True), inplace=True)
train.drop('Cabin', axis=1, inplace=True)
train['Embarked'].fillna(train['Embarked'].value_counts().idxmax(), inplace=True)
train.head()
```

```python
train['Travelled Alone'] = np.where((train["SibSp"] + train["Parch"])>0, 0, 1)
train.drop('Name', axis=1, inplace=True)
train.drop('SibSp', axis=1, inplace=True)
train.drop('Parch', axis=1, inplace=True)
train.drop('Ticket', axis=1, inplace=True)
train.head()
```

```python
training = pd.get_dummies(train, columns=["Pclass","Embarked","Sex"])
training.drop('Sex_female', axis=1, inplace=True)
training.head()
```

```python
test_data = pd.read_csv('C:/Users/Saachi Talwar/Desktop/datasets/titanic_test.csv', delimiter=',')
test_data.head()
```

```python
test = test_data.copy()
test['Age'].fillna(train['Age'].median(skipna=True), inplace=True)
test["Fare"].fillna(train["Fare"].median(skipna=True), inplace=True)
test.drop('Cabin', axis=1, inplace=True)
 
test['Travelled Alone'] = np.where((test["SibSp"] + test["Parch"])>0, 0, 1)
testing = pd.get_dummies(test, columns=["Pclass","Embarked","Sex"])

testing.drop('Name', axis=1, inplace=True)
testing.drop('SibSp', axis=1, inplace=True)
testing.drop('Parch', axis=1, inplace=True)
testing.drop('Ticket', axis=1, inplace=True)
testing.drop('Sex_female', axis=1, inplace=True)

testing.head()
```

```python
train_final = training.copy()
train_final['IsMinor']=np.where(train_final['Age']<=16, 1, 0)
train_final.head()
```

```python
test_final = testing.copy()
test_final['IsMinor']=np.where(test_final['Age']<=16, 1, 0)
test_final.head()
```

```python
clf = LogisticRegression(max_iter=10000)
clf
```

```python
from sklearn.preprocessing import StandardScaler
algo = StandardScaler()
```

```python
X = train_final.drop("Survived",axis=1)
X.head()
```

```python
algo.fit_transform(X)
```

```python
Y = train_final["Survived"]
Y.head(10)
```

```python
clf.fit(X, Y)
```

```python
y_pred = clf.predict(test_final)
y_pred
```

```python
clf.score(X, Y)
```

```python
clf.score(test_final, y_pred)
```

```python
predictions = pd.DataFrame({'Survived':y_pred})
predictions.head()
```

```python
predictions.to_csv('Titanic Predictions', index=False, header=False)
```
